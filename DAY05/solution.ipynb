```python
def digit_sum():
    num = input("Enter a number to calculate sum of it digits")
    sum = 0
    for i in num:
        sum =sum+int(i)
    print(sum)

digit_sum()

```

### 2

```python
def vowel_consonent():
    alpha = input("Enter a alphabet")
    if alpha in ['a','e','i','o','u']:
        print("vowel")
    else:
        print("consonent")

vowel_consonent()
```
