```python
def table():
    number = int(input("enter the number you want to get table for"))
    for i in range(10):
        print(f"{number} X {i} = {number * i}")

table()
```

```python
def calculator():
    number1 = int(input("enter the first number"))
    operator = input("enter the operator +,-,* or /")
    number2 = int(input("Enter the second number"))

    if (operator == '+'):
        print(number1+number2)
    elif (operator == '-'):
        print(number1-number2)
    elif(operator == '*'):
        print(number1*number2)
    else:
        print(number1/number2)

calculator()
```
